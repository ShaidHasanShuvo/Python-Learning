### Why Should We Use **Numpy** instead of **List**?

```python
import numpy as np
import time
import sys

l = range(1000)
print(l)
print(sys.getsizeof(1)*len(l))

a = np.arange(1000)
print(a.size*a.itemsize)

size= 100000
l1= range(size)
l2= np.arange(size)

t1= time.time()
result = [(x+y) for x,y in zip(l1,l2)]
print("python took: ", (time.time()-t1)*1000)

t1= time.time()
result = l1+l2
print("numpy took: ", (time.time()-t1)*1000)

```

###Initialize different types of Arrays

```python
# All Zero's matrix
a= np.zeros((3,4))
print(a)
```

```python
# All One's matrix
b = np.ones((2,3), dtype = 'int32')
print(b)
```

```python
# Any Number Array
a = np.full((3,3), 5, dtype = 'int16')
print(a)
```

```python
# Matrix shape like a
b = np.full_like(a, 12)
print(b)
```

```python
# a & b has same dtype here we can see
b.dtype
```

```python
# Generate Random Decimal Number
np.random.rand(4,2)
```

```python
# random Decimal like a shape
np.random.random_sample(a.shape)
```

```python
# Random int number
np.random.randint(3, 9, size = (3,2))
```

```python
# Indentity matrix
np.identity(4)
```

```python
# Repeat an Array
x = np.array([[5,6,7,8]])
y = np.repeat(x, 4, axis = 0)
print(y)
```

```python
# Challange 1
n = 5

x = np.ones([n,n])
print(x)
x[1:-1,1:-1] = 0
x[2,2]=9
print(x)
```

```python
# Copying an array
a = np.array([1,2,3])
print(a)
b = a.copy()
print(b)

## Don't Do this
c = a
print(c)
c[0] = 5
print(c)
print(a)
```
