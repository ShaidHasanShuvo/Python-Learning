### Basic Graphics

```python
a = [0,1,2]
b= [0,1,4]
# Reshape Figure Size
plt.figure(figsize = (5,3), dpi = 150)

# Ploting a,b
plt.plot(a,b, label = 'X^2', color = 'green')

# Ploting x,y
x = np.arange(0, 6.1, 0.1)
y = np.sin(x)
plt.plot(x, y, label = 'sinx', color = 'red', linestyle= '--', linewidth = 2, marker= '.', markersize=3,
        markeredgecolor= 'blue')

# Title and it's features
plt.title("First Graph", fontdict = {'fontname': 'Comic Sans MS', 'fontsize': 20 })

# Add Labels to X,Y axis
plt.xlabel("x axis")
plt.ylabel("y axis")

# X, Y axis Tickmarks ( Scale your figure)
plt.xticks([0,1,2,3,4,5,6])
plt.yticks([-2,-1,0,1,2,3,4])

# Add Legend
plt.legend()

#Save Your figure (dpi 300 is good for better resulation)
plt.savefig('my_first_graph.png', dpi =300)

# Show plot
plt.show()
```

# Bar Chart

```python
labels = ['A','B','C']
values = [1,4,2]

bars= plt.bar(labels, values)

patterns =['/', '-', '*']
for x in bars:
    x.set_hatch(patterns.pop(0)) # 'pop to give an index for patterns'
#bars[0].set_hatch('/')
#bars[1].set_hatch('-')
#bars[2].set_hatch('*')
plt.figure(figsize=(6,4))

plt.show()
```

# Real World Examples

## Line Chart

```python
gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize= (8,5))

plt.title('Gas Price Over Time (in USD)', fontdict={'fontsize':18, 'fontweight':'bold'})
ax = plt.subplot(111)
for country in gas:
    if country != 'Year':
        plt.plot(gas.Year, gas[country], marker ='.', label = country)
#plt.plot(gas.Year, gas.USA, label = 'USA', 'g.-')
#plt.plot(gas.Year, gas.UK,label = 'UK','r.-')

plt.xticks(gas.Year[::3])
plt.xlabel('Year')
plt.ylabel('Price($)')

ax.legend(bbox_to_anchor =(1,1.02))
plt.show()
```

### Loading the FIFA dataset

## Histogram

```python
fifa = pd.read_csv('fifa_data.csv')
fifa.head()
```

```python
bins =[40,50,60,70,80,90,100]
plt.figure(figsize=(8,5))
plt.hist(fifa.Overall, bins = bins)
plt.title('Distribution of Player by skills in FIFA 2018', fontdict={'fontsize':18,'fontweight':'bold'})

plt.xlabel("Skill Level ")
plt.ylabel('Number of Players')
plt.show()
```

## Pie Chart

```python
fifa.head()
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
Right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

labels = ['Left','Right']
colors = ['#abcdef','#aabbcc']
plt.figure(figsize=(6,6))

plt.pie([left, Right], colors = colors, labels = labels, autopct = '%.2f%%')

plt.title('Players Preferred Foot in FIFA 18')
plt.show()
```

```python
fifa.Weight =  [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]

plt.style.use('ggplot')

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight <= 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight <= 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight <= 200)].count()[0]
heavy =  fifa.loc[fifa.Weight > 200].count()[0]

weights = [light, light_medium,medium,medium_heavy,heavy]
labels =['Under 125', '125 - 150', '150 - 175','175 - 200', 'Above 200'] 
explode = (0.4,.2,0,0,.4)
plt.title('Weight Distribution of Players in FIFA 18')

plt.pie(weights, autopct= '%.2f %%', labels=labels, pctdistance= .8, explode=explode)
plt.show()
```

## Box Plot

```python
plt.style.use('default')
plt.figure(figsize=(6,8))

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
arsenal = fifa.loc[fifa.Club == 'Arsenal']['Overall']

labels = ['FC Barcelona', 'Real Madrid', 'Arsenal']
boxes = plt.boxplot([barcelona, madrid, arsenal], tick_labels = labels, patch_artist= True)

for box in boxes['boxes']:
    # set edge color
    box.set(color= '#abcdef', linewidth = 2)
    #set fill color (patch_artist needed)
    box.set(facecolor= '#e0e0e0')

plt.title('Professional Football Team Comparison')
plt.ylabel('FIFA Overall Ratings')

plt.show()
```
